```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
from plotnine import *

import matplotlib.pyplot as plt
%matplotlib inline 

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors

from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering


from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture

from sklearn.metrics import silhouette_score
from sklearn.mixture import GaussianMixture

%matplotlib inline
```

### **Dataset 1**

```python
x1 = np.array(pd.read_csv("https://raw.githubusercontent.com/cmparlettpelleriti/CPSC392ParlettPelleriti/master/Data/KMEM1.csv"))

plt.scatter(x1[:,0],x1[:,1])
```

# KMeans

```python
dists=[]
for i in range (1,5):

    k=KMeans(n_clusters=i,max_iter=300, tol=0.0001).fit(x1)
    dist= k.inertia_
    dists.append(dist)
plt.plot(range(1,5),dists, marker='o')
```

```python
k=KMeans(n_clusters=2,max_iter=300, tol=0.0001).fit(x1)

labels=k.labels_
score= silhouette_score(x1,labels)
print('score',score)

plt.scatter(x1[:,0],x1[:,1], c=labels)
```

## Agglomerative


```python
Agg=AgglomerativeClustering(n_clusters=2, linkage='average').fit(x1)
labels=Agg.labels_
score= silhouette_score(x1,labels)
print('score',score)

plt.scatter(x1[:,0],x1[:,1], c=labels)
```

## DBSCAN

```python
ds=DBSCAN(eps=0.2,min_samples=2).fit(x1)
labels=ds.labels_
score= silhouette_score(x1,labels)
print('score',score)

plt.scatter(x1[:,0],x1[:,1], c=labels)
```

## GMM

```python
ds=GaussianMixture(n_components=2).fit(x1)
labels=ds.predict(x1)
score= silhouette_score(x1,labels)
print('score',score)

plt.scatter(x1[:,0],x1[:,1], c=labels.astype(float))
```
